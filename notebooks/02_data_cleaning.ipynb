Actually, we only have map data for seven maps, so let's filter that.

```python
map_list = ["de_cache", "de_cbble", "de_dust2", "de_inferno", "de_mirage", "de_overpass", "de_train"]

df3 = df2[df2["map"].isin(map_list)]
df3["map"].value_counts()
```

```python
df3.shape
```

There is no "rank 0" in the rank list according to https://totalcsgo.com/ranks

```python
df4 = df3[(df3["att_rank"] > 0) & (df3["vic_rank"] > 0)]
print(df4["att_rank"].value_counts())
print(df4["vic_rank"].value_counts())
```
